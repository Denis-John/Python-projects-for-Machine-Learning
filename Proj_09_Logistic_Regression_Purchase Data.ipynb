```python
#sigmoid function to map predicted values to probability to probability values
#sigmoid function is used to map predictions to probabilities
import numpy as np # linear Algebra
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("User_Data.csv")
df
```

```python
X=df.iloc[:,[2,3]].values
Y=df.iloc[:,4].values
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
#Feature splitting
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
#Fitting the logistic Regression into the Training Set
from sklearn.linear_model import LogisticRegression
logr=LogisticRegression()
Ypred=logr.fit(Xtrain,Ytrain).predict(Xtest)
```

```python
#ConfusionMatrix
from sklearn.metrics import confusion_matrix
cmat=confusion_matrix(Ypred,Ytest)
```

```python
plt.figure(1)
plt.scatter(X[:,0],X[:,1],c=Y)
plt.suptitle("Purchase Data")
plt.xlabel("Scaled Age")
plt.ylabel("Scaled Income")
plt.grid(1,which="both")
plt.axis("tight")
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
col=sigmoid(np.dot(X,np.transpose(logr.coef_))+logr.intercept_)
#colors=np.random.rand(400)
cf=logr.coef_
xplot=np.arange(-1.0,1.2,0.01)
yplot=-(cf[0,0]*xplot+logr.intercept_)/cf[0,1]
plt.scatter(X[:,0],X[:,1],c=col)
plt.scatter(xplot,yplot)
plt.suptitle("Logistic Regression Purchase Data")
plt.xlabel("Scaled Age")
plt.ylabel("Scaled Income")
plt.grid(1,which="both")
plt.axis("tight")
plt.show()
```
