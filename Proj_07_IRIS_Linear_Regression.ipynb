```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import datasets
import seaborn as sns
import pandas as pd
```

```python
irisset=datasets.load_iris()
```

```python
irisset.keys()
```

```python
irisset.feature_names
```

```python
irisset.data.shape
```

```python
X=irisset.data[:50,:2]
```

```python
reg=LinearRegression().fit(X[:,0:1],X[:,1])
```

```python
#np.reshape(X[:,0],(-1,1))
```

```python
w=reg.coef_
w
```

```python
c=reg.intercept_
```

```python
xpoints=np.linspace(4,6)
ypoints=w[0]*xpoints+c
plt.plot(xpoints,ypoints)
plt.scatter(X[:,0],X[:,1])
plt.suptitle("Linear Regression IRIS Data")  
plt.xlabel("sepal Length")
plt.ylabel(" sepal Width")
plt.grid(1,which="both")
plt.axis("tight")
plt.show()
```

```python
df=pd.DataFrame(X,columns=["sepal Length","sepal Width"])
sns.regplot(x="sepal Length",y="sepal Width",scatter=True,fit_reg=True,data=df)
df
```
